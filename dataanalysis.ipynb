```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import sqlite3
from scipy.stats import ttest_ind
import scipy.stats as stats
warnings.filterwarnings('ignore')
```

```python
numerical_cols=df.select_dtypes(include= np.number).columns

plt.figure(figsize=(15,10))
for i,cols in enumerate(numerical_cols):
    plt.subplot(4,4,i+1)
    sns.histplot(df[cols],kde=True,bins=30)
    plt.title(cols)
plt.tight_layout()
plt.show()
```

```python
for i,cols in enumerate(numerical_cols):
    plt.subplot(4,4,i+1)
    sns.boxplot(df[cols])
    plt.title(cols)
plt.tight_layout()
plt.show()
```

```python
categorical_cols = ['Category', 'OriginCountry']

plt.figure(figsize=(15, 10))
for i, col in enumerate(categorical_cols):
    plt.subplot(1, 2, i + 1)
    sns.countplot(y=df[col], order=df[col].value_counts().index[:10])
    plt.title(f'Countplot of {col}')
plt.tight_layout()
plt.show()
```

```python
corr = df[numerical_cols].corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python

wellness_df = df[df['Category'] == 'Wellness']

wellness_shipping = wellness_df.groupby('OriginCountry')['ShippingCost'].mean().sort_values(ascending=False).reset_index()

# Plotting
plt.figure(figsize=(10,6))
sns.barplot(data=wellness_shipping, x='ShippingCost', y='OriginCountry', palette='coolwarm')

plt.title("Average Shipping Cost for Wellness Products by Country")
plt.xlabel("Average Shipping Cost")
plt.ylabel("Origin Country")
plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats
import numpy as np

# Filter to Wellness products from the USA
usa_wellness = df[(df['OriginCountry'] == 'USA') & (df['Category'] == 'Wellness')]['ShippingCost']

# Sample size, mean, and standard error
n = len(usa_wellness)
mean = np.mean(usa_wellness)
std_err = stats.sem(usa_wellness)  # standard error of the mean

# 95% confidence interval using t-distribution
confidence = 0.95
h = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)

ci_lower = mean - h
ci_upper = mean + h

print(f"Mean Shipping Cost (USA - Wellness): {mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```
